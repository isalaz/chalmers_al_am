```python
from pymongo import MongoClient
import pandas as pd
import os
```

```python

client = MongoClient('localhost', 27017)
```

```python
client.database_names()
```

```python
db = client['in_situ_fluo']
```

```python
db.collection_names()
```

```python
coll = db['scans']
```

```python
items = coll.find()
```

```python
for item in coll.find():
    id = item['_id']
    print(id)
    doc = coll.find_one({'_id' : id})
    print(doc['_id'])
    print(doc['file_path'])
    file_path = doc['file_path']
    fileparts = file_path.split('/')
    scan_str = fileparts[-1]
    new_path = os.path.join(file_path, f'{scan_str}.h5')
    print(new_path)
    coll.update_one({'_id' : id}, {'$set' : {'file_path' : new_path}})
    
```
